```python
#STEP 4: ATR calculations
```

```python
def Wilder(data, periods):
    start = np.where(~np.isnan(data))[0][0] #Check if nans present in beginning
    Wilder = np.array([np.nan]*len(data))
    Wilder[start+periods-1] = data[start:(start+periods)].mean() #Simple Moving Average
    for i in range(start+periods,len(data)):
        Wilder[i] = (Wilder[i-1]*(periods-1) + data[i])/periods #Wilder Smoothing
    return(Wilder)
```

```python
data['prev_close'] = data['CLOSE'].shift(1)
```

```python
data['TR'] = np.maximum((data['HIGH'] - data['LOW']), 
                     np.maximum(abs(data['HIGH'] - data['prev_close']), 
                     abs(data['prev_close'] - data['LOW'])))
```

```python
for i in range(7):
    TR_data = data.iloc[:,i].copy()
    ATR_5 = Wilder(data['TR'], 5)
    ATR_14 = Wilder(data['TR'], 14)
```

```python
list=ATR_5.tolist()
data["ATR_5"]=list
list=ATR_14.tolist()
data["ATR_15"]=list
```

```python
#STEP 5: RSI calculations
```

```python
data['Diff'] = data['CLOSE'].transform(lambda x: x.diff())
data['Up'] = data['Diff']
data.loc[(data['Up']<0), 'Up'] = 0

data['Down'] = data['Diff']
data.loc[(data['Down']>0), 'Down'] = 0 
data['Down'] = abs(data['Down'])

data['avg_5up'] = data['Up'].transform(lambda x: x.rolling(window=5).mean())
data['avg_5down'] = data['Down'].transform(lambda x: x.rolling(window=5).mean())

data['avg_15up'] = data['Up'].transform(lambda x: x.rolling(window=15).mean())
data['avg_15down'] = data['Down'].transform(lambda x: x.rolling(window=15).mean())

data['RS_5'] = data['avg_5up'] / data['avg_5down']
data['RS_15'] = data['avg_15up'] / data['avg_15down']

data['RSI_5'] = 100 - (100/(1+data['RS_5']))
data['RSI_15'] = 100 - (100/(1+data['RS_15']))

data['RSI_ratio'] = data['RSI_5']/data['RSI_15']
```

```python
data=data.drop(['Down', 'avg_5up','avg_5down', 'avg_15up','avg_15down', 'Diff', 'Up',"RS_5","RS_15" ], axis=1)
```

```python
#STEP 6: Stochastic Calculations
```

```python
data['Lowest_15D'] = data["LOW"].transform(lambda x: x.rolling(window = 15).min())
data['High_15D'] = data['HIGH'].transform(lambda x: x.rolling(window = 15).max())

data['Stochastic_15'] = ((data['CLOSE'] - data['Lowest_15D'])/(data['High_15D'] - data['Lowest_15D']))*100

data['Stochastic_%K_15'] = data['Stochastic_15'].rolling(window = 15).mean()
```

```python
data['Stochastic_%D'] = data['Stochastic_%K_15'].transform(lambda x: x.rolling(window = 3).mean())
```

```python
data = data.drop(['Lowest_15D', 'High_15D','Stochastic_15'], axis=1)
```

```python
#Method: Use Pearson's R and whatever hours lag gives the highest correlation, it has that
#many hours delayed effect on the stock prices
```

```python
lag=[]

for i in range(21):
    lag.append(i)
```

```python
#ATRA - NOTE: 10 periods of delay were inspected - 15 periods long indicators yielded error
#So 20 periods lag is created
```

```python
data=data.dropna()
```

```python
#Dropped NaN to avoid any mistakes - though I'm not sure about it
```

```python
y1=data['SMA_ratio']
y2=data['CLOSE']

list=[]

df = pd.DataFrame(y1)

for i in range(21):
    y2=y2.shift(periods=i)
    df2=pd.DataFrame(y2)
    final=pd.concat([df,df2], axis=1)
    corr=final.corr()
    item=corr.iloc[0,1]
    list.append(item)
    
df=pd.DataFrame({"Lag":lag, 
                 "SMA_ratio Correlation":list})

df.sort_values(by='SMA_ratio Correlation', key=pd.Series.abs, ascending=False)
```

```python
#Conclusion: SMA_ratio has no delayed effect
```
